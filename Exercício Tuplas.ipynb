### Maior e menor valores em Tupla

```python
from random import randint
numeros = (randint(1, 10), randint(1, 10), randint(1, 10), randint(1, 10), randint(1, 10), randint(1, 10))
print('Os valores Sorteados foram:  ', end='')
for n in numeros:
    print(f'{n} ', end='')
print(f'\nO maior valor sorteado foi o {max(numeros)}')
print(f'\nO menor valor sorteado foi o {min(numeros)}')
```

## Exercicio Análise de dados em uma Tupla

```python
num = (int(input('Digite o 1° número: ')),
      int(input('Digite o 2° número: ')),
      int(input('Digite o 3° número: ')),
      int(input('Digite o 4° número: ')))

print(f'Você digitou os números {num}')
print(f'O maior número digitado foi o {max(num)}')
print(f'O menor número digitado  foi o {min(num)}')
print(f'O valor 9 apareceu {num.count(9)} vezes')
if 3 in num:
    print(f'O valor 3 apareceu na {num.index(3) + 1}° posição')
else:
    print('Você não digitou o número 3 em nenhuma posição!')
print('\nOs números pares digitados foram ', end='')
for n in num:
    if n % 2 == 0:
        print(n, end=' ')
```

## Exercicio  Lista de Preços com Tupla

```python
listagem = ('Lápis' , 1.75,
           'Borracha' , 2,
           'Caderno' , 15.35,
           'Estojo' , 40,
           'Transferidor' , 5.10,
           'Compasso' , 6.60,
           'Mochila' , 130,
           'Livro' , 60.49,)
print('-' * 50)
print(f'{"LISTAGEM DE PREÇOS":^40}')
print('-' * 50)
for pos in range(0, len(listagem)):
    if pos % 2 == 0:
        print(f'{listagem[pos]:.<43}', end='')
    else:
        print(f'R${listagem[pos]:.>4.2f}')
print('-' * 50)
```

## Exercicio Contando vogais em Tupla

```python
palavras = ('APRENDER',
           'PROGRAMAR',
           'LINGUAGEM',
           'PHYTON',
           'CURSO',
           'GRATIS',
           'ESTUDAR',
           'PRATICAR',
           'TRABALHAR',
            'MERCADO',
            'PROGRAMADOR',
            'FUTURO')
for p in palavras:
    print(f'\nNa palavra {p.upper()} temos as vogais: ', end='')
    for letra in p:
        if letra.lower() in 'aeiou':
            print(letra, end=' ')
           
```
